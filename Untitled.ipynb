```python
Image_size = (100,100)
```

```python
validation_directory_dataset = validation_directory_dataset.map(lambda x, y: (tensorflow.image.resize(x, Image_size), y))
validation_directory_dataset = validation_directory_dataset.map(lambda x, y: (x / 255.0, y))
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import RandomRotation, RandomZoom, RandomContrast, RandomFlip
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import RandomRotation, RandomZoom, RandomContrast, RandomFlip

# Define a data augmentation layer sequence directly in your model
data_augmentation = Sequential([
    RandomRotation(factor=0.1),  # Randomly rotate images in the range (0 to 0.1 * 2)
    RandomZoom(height_factor=0.2, width_factor=0.2),  # Randomly zoom images in and out by 20%
    RandomContrast(factor=0.2),  # Randomly change image contrast by 20%
    RandomFlip("horizontal")  # Randomly flip images horizontally
])
```

```python
class_names = train_directory_dataset.class_names
```

```python
plt.figure(figsize=(15, 10))
for images, labels in train_directory_dataset.take(1):
    for i in range(9):
        augmented_image = data_augmentation(images)
        plt.subplot(3, 3, i + 1)
        plt.imshow(augmented_image[0].numpy().astype("uint8"))
        plt.title("Class: " + class_names[labels[0]])
        plt.axis("off")
plt.show()
```

```python
keras.backend.clear_session()
tensorflow.random.set_seed(42)
np.random.seed(42)

base_model_mobilenet = tensorflow.keras.applications.MobileNetV2(
                                    weights="imagenet", include_top=False)
```

```python
inputs = keras.Input(shape=(100, 100, 3))
x = data_augmentation(inputs)
x = base_model_mobilenet(x, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(128, activation='relu')(x)
x = layers.Dropout(0.5)(x)
outputs = layers.Dense(number_of_classes, activation="softmax")(x)
mobilenet_model = Model(inputs=inputs, outputs=outputs)
```

```python
for layer in base_model_mobilenet.layers:
    layer.trainable = False
```

```python
mobilenet_model.compile(optimizer=keras.optimizers.Adam(lr=1e-4),
              loss=keras.losses.SparseCategoricalCrossentropy(),
              metrics=["accuracy"])
mobilenet_model.summary()
```

```python
early_stopping_cb = tensorflow.keras.callbacks.EarlyStopping(
    patience=3, # Number of epochs with no improvement after which training will be stopped.
    restore_best_weights=True, # Whether to restore model weights from the epoch with the best value of the monitored quantity.
)
from tensorflow.keras.callbacks import TensorBoard
import datetime

# Define the log directory for TensorBoard
log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")

# Create a TensorBoard callback
tensorboard_cb = TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
epochs=40
history_mobilenet = mobilenet_model.fit(train_directory_dataset, epochs=epochs, 
                                        
                                        validation_data=validation_directory_dataset, 
                                        callbacks=[early_stopping_cb, tensorboard_cb])
```
